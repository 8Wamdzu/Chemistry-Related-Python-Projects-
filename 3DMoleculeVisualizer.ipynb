```python
#aight so you already know we had to get all da imports

from unittest import result
from rdkit import Chem
from rdkit.Chem import AllChem
import py3Dmol
import pubchempy as pcp
from chempy import Substance

#now we gotta make a function that converts the users input into a SMILES string you feel me?
#SMILES stands for Simplified Molecular Input Line Entry System, 
#and it's how the computer boutta read the input 

def get_smiles_from_input(molecule):
    results = pcp.get_compounds(molecule, 'name')
    if results:
       
        return results[0].smiles 
    else:
        return "I ain't gon lie this one not in the database or you prolly spelled it wrong."
    
#then we gotta ask the user for the molecule

molecule = input("Please type name: ")
smiles = get_smiles_from_input(molecule)

#now we gotta get the 2D coordinates using RDkit. The Chem.AddHs adds the implicit hydrogrens 
# #hydrogenlivesmatter #stopleavingouthydrogen 
imperfectmol = Chem.MolFromSmiles(smiles)
imperfectmol = Chem.AddHs(imperfectmol)
AllChem.EmbedMolecule(imperfectmol
                      )

perfectmol = Chem.MolToMolBlock(imperfectmol)

viewer = py3Dmol.view(width=400, height=400)
viewer.addModel(perfectmol, 'mol')
viewer.setStyle({'stick': {}, 'sphere': {'radius': 0.42}})
viewer.addStyle({'elem': 'C'}, {'stick': {'color': '#333333'}}) 
viewer.addStyle({'elem': 'N'}, {'stick': {'color': '#0000FF'}}) 
viewer.addStyle({'elem': 'O'}, {'stick': '#FF0000'})            

viewer.zoomTo()
viewer.show()




  





```
